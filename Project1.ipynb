# Step 6: Plot the Data

```python
x = ['English', 'French', 'Italian', 'German','Spanish','pt-br']
y = [4.3, 5.2, 6.1, 5.0, 5.3, 5.3]

x1 = x[0]
y1 = y[0]
plt.bar(x1, y1, label = "51000-60000") 
  
x2 = x[1]
y2 = y[1]
plt.bar(x2, y2, label = '41000-50000') 

x3 = x[2]
y3 = y[2]
plt.bar(x3, y3, label = '31000-40000') 

x4 = x[3]
y4 = y[3]
plt.bar(x4, y4, label = '21000-30000') 

x5 = x[4]
y5 = y[4]
plt.bar(x5, y5, label = '11000-20000') 

x6 = x[5]
y6 = y[5]
plt.bar(x6, y6, label = '0-10000') 

plt.xticks(rotation=90)
# naming the x axis 
plt.xlabel('Language') 
# naming the y axis 
plt.ylabel('Privacy Average') 
  
# show a legend on the plot 
ax = plt.subplot(111)
chartBox = ax.get_position()
ax.set_position([chartBox.x0, chartBox.y0, chartBox.width*0.6, chartBox.height])
ax.legend(loc='upper center', bbox_to_anchor=(1.45, 0.8), shadow=True, ncol=1) 
  
# function to show the plot 
plt.show() 
```
